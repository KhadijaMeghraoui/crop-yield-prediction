**COMBINING USING SIGMOID**

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate weights using sigmoid function
weights_X1 = sigmoid(X1)
weights_X2 = sigmoid(X2)

# Normalize weights
total_weight = weights_X1 + weights_X2
weights_X1 /= total_weight
weights_X2 /= total_weight

print("Weights using Sigmoid:")
print("Weight for X1 predictions:", weights_X1)
print("Weight for X2 predictions:", weights_X2)

# Combine predictions using weighted sum
combined_predictions = weights_X1 * X1 + weights_X2 * X2

print("Combined Predictions:", combined_predictions)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y, combined_predictions)
mae
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, combined_predictions)
```
