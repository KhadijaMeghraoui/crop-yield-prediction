**APPLYING AUTOENCODERS WITH DNN**

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
import tensorflow as tf
```

```python
# Define the autoencoder model
def build_autoencoder(input_dim, encoding_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(encoding_dim, activation='relu')(input_layer)
    decoded = Dense(input_dim, activation='sigmoid')(encoded)
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

# Define the predictive model
def build_predictive_model(encoding_dim, units, learning_rate):
    model = tf.keras.Sequential([
        Dense(units, activation='relu', input_dim=encoding_dim),
        Dense(1, activation='linear')
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mse')
    return model

# Hyperparameter search space
param_grid = {
    'encoding_dim': [20, 50, 70, 80],
    'units': [32, 64, 128],
    'learning_rate': [0.01, 0.001, 0.0001]
}

# Perform grid search
best_score = float('inf')
best_params = {}

for encoding_dim in param_grid['encoding_dim']:
    for units in param_grid['units']:
        for learning_rate in param_grid['learning_rate']:
            # Build and train the autoencoder
            autoencoder = build_autoencoder(input_dim=100, encoding_dim=encoding_dim)
            autoencoder.fit(X_train, X_train, epochs=20, batch_size=32)

            # Extract encoder from the trained autoencoder
            encoder = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer(index=-2).output)

            # Encode the data using the trained encoder
            X_train_encoded = encoder.predict(X_train)
            X_test_encoded = encoder.predict(X_test)

            # Build and train the predictive model
            predictive_model = build_predictive_model(encoding_dim=encoding_dim, units=units, learning_rate=learning_rate)
            predictive_model.fit(X_train_encoded, y_train, epochs=20, batch_size=32)

            # Evaluate the model on the test set
            score = predictive_model.evaluate(X_test_encoded, y_test)

            # Check if this set of hyperparameters improves the model
            if score < best_score:
                best_score = score
                best_params = {
                    'encoding_dim': encoding_dim,
                    'units': units,
                    'learning_rate': learning_rate
                }

print("Best Hyperparameters:", best_params)
print("Best Score (MSE):", best_score)
```

```python
# Best hyperparameters
best_encoding_dim = best_params['encoding_dim']
best_units = best_params['units']
best_learning_rate = best_params['learning_rate']
```

```python
# Build and train the autoencoder
autoencoder = build_autoencoder(input_dim=100, encoding_dim=best_encoding_dim)
autoencoder.fit(X_train, X_train, epochs=20, batch_size=32)
```

```python
# Extract encoder from the trained autoencoder
encoder = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer(index=-2).output)
```

```python
# Encode the data using the trained encoder
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)
```

```python
# Build and train the predictive model
predictive_model = build_predictive_model(encoding_dim=best_encoding_dim, units=best_units, learning_rate=best_learning_rate)
predictive_model.fit(X_train_encoded, y_train, epochs=20, batch_size=32)
```

```python
# Evaluate the model on the test set
evaluation = predictive_model.predict(X_test_encoded)
```

**APPLYING ADABOOST WITH HYPERPARAMETERS**

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

# Hyperparameter search space for Adaboost
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0]
}

# Base estimator for Adaboost
base_estimator = DecisionTreeRegressor()

# Create Adaboost model
adaboost_model = AdaBoostRegressor(base_estimator=base_estimator)

# Perform grid search
grid_search = GridSearchCV(adaboost_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
best_score = -grid_search.best_score_

```

```python
final_adaboost_model = AdaBoostRegressor(base_estimator=base_estimator, **best_params)
final_adaboost_model.fit(X_train, y_train)
```

```python
y_adaboost = final_adaboost_model.predict(X_test)
```

```python
m3=mean_absolute_error(y_test, y_adaboost)
m3
```

```python
m31=mean_squared_error(y_test, y_adaboost)
m31
```

```python
rmse31= np.sqrt(mean_squared_error(y_test, y_adaboost))
rmse31
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_adaboost, label='AdaBoost', alpha=0.7, color='y')

plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Evaluation of Model Predictions')
plt.legend()

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from skopt import BayesSearchCV
from skopt.space import Real, Integer, Categorical
```

```python
# Define the search space for hyperparameters
search_space = {
    'n_estimators': Integer(50, 500),
    'max_depth': Integer(10, 50),
    'min_samples_split': Integer(2, 20),
    'min_samples_leaf': Integer(1, 10),
    'max_features': Categorical(['log2', 'sqrt']),
    'max_leaf_nodes': Integer(10, 50),
    'min_impurity_decrease': Real(0.0, 0.2, prior='uniform')
}
```

```python
def rf_objective(params):
    rf = RandomForestRegressor(**params, random_state=42)
    score = -np.mean(cross_val_score(rf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    return score
```

```python
# Create the BayesSearchCV object
bayes_search = BayesSearchCV(
    RandomForestRegressor(),
    search_space,
    n_iter=10,  # Number of iterations for Bayesian optimization
    cv=5,       # Number of cross-validation folds
    n_jobs=-1,
    random_state=42
)

# Perform Bayesian optimization
bayes_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = bayes_search.best_params_
print("Best hyperparameters:", best_params)
```

```python
# Train the final model with the best hyperparameters
model_RF = RandomForestRegressor(**best_params, random_state=42)
model_RF.fit(X_train, y_train)
```

```python
ypred_RF = model_RF.predict(X_test)
```
